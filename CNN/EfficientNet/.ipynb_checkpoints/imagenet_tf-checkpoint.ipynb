```python
def build_learning_rate(initial_lr,
                        global_step,
                        steps_per_epoch=None,
                        lr_decay_type='exponential',
                        decay_factor=0.97,
                        decay_epochs=2.4,
                        total_steps=None,
                        warmup_epochs=5):
    """Build learning rate."""
    if lr_decay_type == 'exponential':
        assert steps_per_epoch is not None
        decay_steps = steps_per_epoch * decay_epochs
        lr = tf.compat.v1.train.exponential_decay(
            initial_lr, global_step, decay_steps, decay_factor, staircase=True)
    elif lr_decay_type == 'cosine':
        assert total_steps is not None
        lr = 0.5 * initial_lr * (
            1 + tf.cos(np.pi * tf.cast(global_step, tf.float32) / total_steps))
    elif lr_decay_type == 'constant':
        lr = initial_lr
    else:
        assert False, 'Unknown lr_decay_type : %s' % lr_decay_type

    if warmup_epochs:
        warmup_steps = int(warmup_epochs * steps_per_epoch)
        warmup_lr = (
            initial_lr * tf.cast(global_step, tf.float32) / tf.cast(
                warmup_steps, tf.float32))
        lr = tf.cond(global_step < warmup_steps, lambda: warmup_lr, lambda: lr)

    return lr


def build_optimizer(learning_rate,
                    optimizer_name='rmsprop',
                    decay=0.9,
                    epsilon=0.001,
                    momentum=0.9):
    """Build optimizer."""
    if optimizer_name == 'sgd':
        optimizer = tf.train.GradientDescentOptimizer(learning_rate=learning_rate)
    elif optimizer_name == 'momentum':
        optimizer = tf.train.MomentumOptimizer(
        learning_rate=learning_rate, momentum=momentum)
    elif optimizer_name == 'rmsprop':
        optimizer = RMSprop(lr=learning_rate, decay=decay, epsilon=epsilon)
    else:
        logging.fatal('Unknown optimizer: %s', optimizer_name)

    return optimizer
```

```python
train_dir = '/media/data2/dataset/imagenet/train/'
val_dir = '/media/data2/dataset/imagenet/val/'

init_lr = 0.256
momentum = 0.9
bn_momentum = 0.99
weight_decay = 1e-5

batch_size = 128
epochs = 100
```

```python
model = EfficientNetB0(weights=None)
```

```python
model.summary()
```

```python
train_data = ImageDataGenerator(preprocessing_function=preprocess_input)
val_data = ImageDataGenerator(preprocessing_function=preprocess_input)
```

```python
train_generator = train_data.flow_from_directory(train_dir,
                                                 target_size=(224, 224),
                                                 batch_size=batch_size,
                                                 shuffle=True,
                                                 class_mode='categorical')

val_generator = train_data.flow_from_directory(val_dir,
                                                 target_size=(224, 224),
                                                 batch_size=batch_size,
                                                 shuffle=False,
                                                 class_mode='categorical')
```

```python
lr = build_learning_rate(init_lr, epochs, len(train_generator))
optimizer = build_optimizer(lr)

model.compile(optimizer=optimizer,
              loss='categorical_crossentropy',
              metrics=['accuracy'])

print(len(model.trainable_weights))
```

```python
history = model.fit_generator(train_generator,
                            steps_per_epoch=len(train_generator),
                            epochs=epochs,
                            validation_data=val_generator,
                            validation_steps=len(val_generator))
```
